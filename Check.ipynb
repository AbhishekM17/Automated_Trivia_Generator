```python
from moviepy.video.io.VideoFileClip import VideoFileClip

def split_video(input_file, segment_duration=5, start_time=10):
    # Load the video
    video = VideoFileClip(input_file)
    video_duration = video.duration

    # Calculate number of segments starting from start_time
    num_segments = int((video_duration - start_time) // segment_duration)

    for i in range(num_segments):
        segment_start = start_time + i * segment_duration
        segment_end = segment_start + segment_duration
        
        # Extract and save each segment
        segment = video.subclip(segment_start, segment_end)
        output_file = f"output_{i + 1:03d}.mp4"  # output format as mp4 for compatibility
        segment.write_videofile(output_file, codec="libx264", audio_codec="aac")
    
    video.close()

# Example usage with a .webm file
split_video("videoplayback.webm")
```

```python
import ffmpeg

def split_video(input_file, segment_duration=5, start_time=10):
    # Get video duration
    probe = ffmpeg.probe(input_file, v='error', select_streams='v:0', show_entries='format=duration')
    video_duration = float(probe['format']['duration'])

    # Calculate number of segments starting from start_time
    num_segments = int((video_duration - start_time) // segment_duration)

    for i in range(num_segments):
        segment_start = start_time + i * segment_duration
        segment_end = segment_start + segment_duration
        output_file = f"output_{i + 1:03d}.webm"  # output as .webm

        # Use FFmpeg to cut the video from start_time to end_time
        ffmpeg.input(input_file, ss=segment_start, t=segment_duration).output(output_file).run()

# Example usage
split_video("videoplayback.webm")
```

```python
from moviepy.editor import VideoFileClip

# Load the original .webm video
video = VideoFileClip("videoplayback.webm")

# Start time of the first segment
start_time = (10*60) + 35

# Segment duration (6.4 seconds and 3.6 seconds)
first_part_duration = 6.4
second_part_duration = 3.6

# Counter for video numbering
video_count = 3

# Set the maximum number of segments
max_video_count = 3  # Adjust this to the number of segments you want

# Split the video into segments until the end or until the max count is reached
while start_time < video.duration and video_count <= max_video_count:
    # Calculate the end time for the first part
    part_1_end = min(start_time + first_part_duration, video.duration)

    # Extract the first part
    part_1 = video.subclip(start_time, part_1_end)
    part_1.write_videofile(f"segment_{video_count}_part_1_.webm", codec="libvpx", audio_codec="libvorbis")

    # Update the start time for the second part
    start_time = part_1_end
    part_2_end = min(start_time + second_part_duration, video.duration)

    # Extract the second part
    part_2 = video.subclip(start_time, part_2_end)
    part_2.write_videofile(f"segment_{video_count}_part_2_.webm", codec="libvpx", audio_codec="libvorbis")

    # Update the start time for the next segment
    start_time = part_2_end

    # Increment the video count
    video_count += 1
```
